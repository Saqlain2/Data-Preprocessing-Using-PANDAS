# B]  ORDINAL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r"C:\Users\SaifzWorld.Lenovo-PC\Desktop\saqlain_csv\house_pred.csv")
```

```python
df
```

will work on 2 columns only

```python
df2= df[["KitchenQual","BldgType"]]
```

now label encoder

```python
le =LabelEncoder()
```

```python
le.fit_transform(df2["BldgType"])
```

```python
df2["BldgType_le"] = le.fit_transform(df2["BldgType"])
```

as we can see up label counts gives importance to number first then alphabet 1,2,3 and then a,b,c
0 = 1fam
1 = 2fmcon
2 = duplex

now ordinal encoding 

```python
df2["KitchenQual"].value_counts()
```

EX = EXCELLENT
TA = TYPICAL
FA = FAIR
GD = GOOD

now will create a dict for ordinal encoding 


```python
ordinal_enc = {'Ex':4,'Gd':3,'Fa':2,'TA':1}
```

```python
df2["KitchenQual_ord"]=df2["KitchenQual"].map(ordinal_enc)
```
