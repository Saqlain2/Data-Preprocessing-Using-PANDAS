```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:\\Users\\SaifzWorld.Lenovo-PC\\Desktop\\saqlain_csv\\labeledTrainData.tsv",delimiter = "\t")
```

```python
data.tail()
```

```python
np.bincount(data.sentiment)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()


x_train,x_test,y_train,y_test=train_test_split(data.review,data.sentiment,test_size=0.3)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)
```

```python
features_name = vectorizer.get_feature_names()
print("number of features: {}".format(len(features_name)))
```

```python
vectorizer.vocabulary_
```

```python
#CV AND ACCURACY
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
scores =cross_val_score(LogisticRegression(),x_train,y_train,cv=5)
print("mean cv accuracy : {}".format(np.mean(scores)))
```

```python
#Checking LOG REG ,NBMULTINOMIAL ,RANDOM FOREST
```

```python
logregression = LogisticRegression()
logregression.fit(x_train,y_train)
print("training set scores : {}".format(logregression.score(x_train,y_train)))
print("test set scores : {}".format(logregression.score(x_test,y_test)))
```

```python
from sklearn.metrics import confusion_matrix
pred_log = logregression.predict(x_test)
confusion_mat = confusion_matrix(y_test,pred_log)
print("confusion mat :\n{}".format(confusion_mat))
```

```python
from sklearn.naive_bayes import MultinomialNB
nb =  MultinomialNB()
nb.fit(x_train,y_train)
print("training set scores : {}".format(nb.score(x_train,y_train)))
print("test set scores : {}".format(nb.score(x_test,y_test)))
```

```python
pred_log = nb.predict(x_test)
confusion_mat = confusion_matrix(y_test,pred_log)
print("confusion mat :\n{}".format(confusion_mat))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf =  RandomForestClassifier()
rf.fit(x_train,y_train)
print("training set scores : {}".format(rf.score(x_train,y_train)))
print("test set scores : {}".format(rf.score(x_test,y_test)))
```

```python
pred_log = rf.predict(x_test)
confusion_mat = confusion_matrix(y_test,pred_log)
print("confusion mat :\n{}".format(confusion_mat))
```
